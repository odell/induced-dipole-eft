```python
import numpy as np
import matplotlib.pyplot as plt

import constants as const

plt.style.use('science')
gr = 4/3
h = 3
```

```python
spectra_lo = np.loadtxt('spectra_local_lo_a0_fix.txt')
spectra_mm = np.loadtxt('spectra_Mott_Massey.txt')
```

```python
spectra_lo[:, 1]
```

```python
ii = np.where(const.BETA4/spectra_lo[:, 0] < 31)[0]
```

```python
fig, ax = plt.subplots(figsize=(gr*h, h), dpi=300)
fig.patch.set_facecolor('white')

for i in range(1, 8):
    ax.plot(const.BETA4/spectra_lo[ii, 0], -spectra_lo[ii, i], color='C0')

for e in spectra_mm:
    ax.axhline(-e, linestyle='--', color='C6')

ax.set_yscale('log')
ax.set_xlabel(r'$\beta_4/R$')
ax.set_ylabel(r'$B_n$ (a.u.)')

plt.savefig('figures/lo_spectrum_R.pdf')
```
